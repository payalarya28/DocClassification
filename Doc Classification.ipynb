<h1>Unsupervised Document Classification / Document Clustering</h1>

# Objective of this notebook

### Cluster similar groups of documents together in an unsupervised way
    
We have around 500 word documents which we need to cluster in different categrories. We are using LDA model to achieve this.
LDA states that each document in a corpus is a combination of a fixed number of topics. A topic has a probability of generating various words, where the words are all the observed words in the corpus. These ‘hidden’ topics are then surfaced based on the likelihood of word co-occurrence.

<h4>Test the model for different topic numbers to get highest coherence value</h4>

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        tfidf = models.tfidfmodel.TfidfModel(corpus)
        transformed_tfidf = tfidf[corpus]
        model = models.ldamulticore.LdaMulticore(corpus=transformed_tfidf, num_topics=num_topics, id2word=dictionary)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python

model_list, coherence_values = compute_coherence_values(dictionary=dictionary_bi, corpus=corpus_bi, texts=tokenized_text_bi, start=3, limit=10, step=1)
```

```python

# Show graph
limit=10; start=3; step=1;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()# Print the coherence scores
```

```python
# Print the coherence scores
for m, cv in zip(x, coherence_values):
    print("Num Topics =", m, " has Coherence Value of", round(cv, 4))
```
